```python
import requests
from bs4 import BeautifulSoup as bs
```

```python

def process_document(document):
    
    nodes = document.select('#breaknews_body > dl')
    data = list()
    links = list()


    # check if is empty element

    for li in nodes[0].find_all('a', href=True):
        li_link = li['href']#get links
        if(li_link not in links):#avoid finding the links have found
            print(li_link)
            print("-----")
            # get into links
            li_res = requests.get(li_link)
            li_doc = bs(li_res.text, 'lxml')

            # get content
            li_content = ""
            for ele in li_doc.select('div#story_body_content > p'):
                li_content += ele.get_text()

            # append new row
            data.append({
                'link' : li_link,
                'content' : li_content,
            })
            #put the link have found into list
            links.append(str(li_link))

        
    return data
```

```python
cnt = 0
all_data = list()
res = requests.get('https://udn.com/rank/pv/2/6640')
doc = bs(res.text, 'html.parser')
#print(doc)
all_data = process_document(doc)
```
